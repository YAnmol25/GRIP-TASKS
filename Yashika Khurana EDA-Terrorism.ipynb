### Missing values through out columns seen as above

```python
regions=np.unique(df['region_txt'], return_counts=True)
```

```python
print(regions)
```

```python
len(regions[0])
```

```python
idx=np.argmax(regions[1])
max_region=regions[0][idx]
print(max_region)
```

### Data of 12 regions is present.  Maximum events have been observed in Middle East & North Africa.

### Let's confirm it visually

```python
plt.figure(figsize=(30,20))
sns.countplot(x="region_txt", data=df)
plt.xticks(
  rotation=45, 
   horizontalalignment='right',
   fontweight='light',
     fontsize='x-large'  
)
```

### Thus confirmed.

```python

top_countries = df.country_txt.value_counts().reset_index().rename(columns={'index': 'Country Name', 'country_txt':'Count'})
```

```python
top_10=top_countries.head(10)
```

```python
top_10
```

```python
plt.figure(figsize=(12,10))
sns.barplot(x="Country Name", y="Count",data=top_10)
```

### Most of the events have been observed in Iraq, followed by Pakistan, Afghanistan, India etc.

```python
top_targets = df.targtype1_txt.value_counts().reset_index().rename(columns={'index': 'Target', 'targtype1_txt':'Count'})
```

```python
top_targets
```

```python
plt.figure(figsize=(10,10))
plt.pie(top_targets["Count"], labels=top_targets["Target"])
```

### Most targetted places are private citizens & property , followed by military, police, Government etc.

```python
years = df.iyear.value_counts().reset_index().rename(columns={'index': 'Year', 'iyear':'Count'})
```

```python
years
```

### Year 2014 accounts for most of the cases

```python
weapons = df.weaptype1_txt.value_counts().reset_index().rename(columns={'index': 'Weapon', 'weaptype1_txt':'Count'})
```

```python
weapons
```

```python
plt.figure(figsize=(20,20))
plt.bar(weapons["Weapon"],weapons["Count"], width=0.3, alpha=0.5 )
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'  
)
```

### Explosives are the most used weapons, followed by firearms, unknown, etc.
