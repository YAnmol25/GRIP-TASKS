### In both the tables, team- Rising Pune Supergiant & Rising Pune Supergiants have been mentioned. We need to convert it to one only

```python
matches=matches.replace('Rising Pune Supergiants','Rising Pune Supergiant')
deliveries=deliveries.replace('Rising Pune Supergiants','Rising Pune Supergiant')
```

```python
plt.figure(figsize=(20,20))
sns.countplot(x="toss_winner", data=matches)
plt.xticks(
  rotation=45, 
   horizontalalignment='right',
   fontweight='light',
     fontsize='x-large'  
)
```

### Most tosses have been won by Mumbai Indians

```python
plt.figure(figsize=(50,50))
sns.countplot(matches["winner"], data=matches)
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'  
)
```

### Most matches have been won by MI, followed by CSK

```python
td=np.unique(matches['toss_decision'], return_counts=True)
plt.pie(td[1], labels=td[0], autopct="%f")
```
