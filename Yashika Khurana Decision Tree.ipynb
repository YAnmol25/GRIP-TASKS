## <span style='color:Blue'> Importing library files  </span>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

## <span style='color:Blue'> Loading dataset </span>

```python
iris=datasets.load_iris()
x=pd.DataFrame(iris.data, columns=iris.feature_names)
```

## <span style='color:Blue'> Processing data</span>

```python
x.head()
```

```python
y=pd.DataFrame(iris.target,columns=["Species"])
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

## <span style='color:Blue'> Creating decision trees </span>

```python
dt= DecisionTreeClassifier() # uses gini
```

```python
dt.fit(x_train,y_train)
```

```python
dt.predict(x_test)
```

```python
y_test
```

## <span style='color:Blue'> Evaluating the decision trees </span>

```python
dt.score(x_test,y_test)
```

### Score: 1 (with gini)

```python
dt2= DecisionTreeClassifier(criterion='entropy',max_depth=5) # uses entropy
```

```python
dt2.fit(x_train,y_train)
```

```python
dt2.score(x_test,y_test)
```

### Score: 1 (with entropy)

## <span style='color:Blue'> Predictions on testing data </span>

## <span style='color:Blue'> Visualising decision tree </span>

```python
from sklearn import tree
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(dt, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```
